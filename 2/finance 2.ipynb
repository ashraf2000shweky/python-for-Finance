```python
import pandas as pd
import yfinance as yf # yahoo finance مكتبة لتحميل الأوراق المالية من موقع  
import datetime as dt
import pandas_datareader as pdr
import matplotlib.pyplot as plt
```

```python
ticker = 'AAPL'
start = dt.datetime(2021,1,1)
end = dt.datetime(2022,1,1)
data = yf.download('AAPL',start=start,end=end)  # yahoo تحميل سوق الورق المالي من 
```

```python
# إظهار الشكل البياني بشكل أفضل 
%matplotlib notebook 
```

```python
fig , ax = plt.subplots() # إنشاء مجموعة أشكال بيانية
data['Close'].plot(ax=ax)
ax.set_ylabel('Price')
ax.set_title('AAPL')
```

```python
fig ,ax = plt.subplots(2,2)
data['Open'].plot(ax=ax[0,0],title='Open')
data['High'].plot(ax=ax[0,1],title='High')
data['Low'].plot(ax=ax[1,0],title='Low')
data['Close'].plot(ax=ax[1,1],title='Close')
plt.tight_layout()
```

```python
fig,ax = plt.subplots()
data['Volume'].loc['2021-07-01':'2021-08-15'].plot.barh(ax=ax)
plt.tight_layout()
```
