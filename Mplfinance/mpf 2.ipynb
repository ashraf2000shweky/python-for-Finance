```python
added_plots =  [ mpf.make_addplot(data['High']),
                mpf.make_addplot(data['Low']),
                mpf.make_addplot(data['Open']),
                mpf.make_addplot(data['Close']) 
               ]



fig, axes = mpf.plot(data, type="candle",
                     addplot=added_plots,
                     volume=True,
                     returnfig=True)
                     

axes[0].set_ylabel("AAPL")
axes[2].set_ylabel("Volume AAPL")
axes[0].legend(['_nolegend_','_nolegend_','High','Low','Open','Close'])
plt.show()
```

```python
data['macd'], data['macdsignal'], data['macdhist'] =tb.MACD(data['Close'], fastperiod=12, slowperiod=26,signalperiod=9)
```

```python
fig , ax =plt.subplots(3,1,sharex=True,figsize=(10,10),gridspec_kw= {'height_ratios':[3,2,2]})

fig.set_edgecolor('k')
fig.set_linewidth(4)
fig.set_facecolor('cyan')
ax3 = ax[2].twinx()

p2 = [ mpf.make_addplot(data['High'],color='g',ax=ax[2]),  
       mpf.make_addplot(data['Low'],color='b',ax=ax[2]),
       mpf.make_addplot(data['macd'],ax=ax3,color='r')] 
                         
                         
mpf.plot(data,type='candle',ax=ax[0], addplot=p2 ,style='blueskies',volume=ax[1],mav=(6,9))
fig.suptitle('AAPL')

ax[0].legend(labels=['_nolegend_','_nolegend_','mav6','mav9'])
ax[2].legend(['High','Low'])
ax3.legend(['MACD'],bbox_to_anchor=[0.7,0.2])
ax[2].grid(False)
ax[0].set_ylabel('Price',fontsize=15,labelpad=30,color='r')
ax[1].set_ylabel('Volume',fontsize=10,labelpad=35,color='r')

plt.tight_layout()
plt.show()
```
